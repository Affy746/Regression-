(1) What is Simple Linear Regression?

Ans: A model in which there is one dependent variable and only one independent variable, is called as simple linear regression.





(2) What are the key assumptions of Simple Linear Regression?

Ans: Assumptions are as follows;
Linearity: Relationship between two variables must be linear.
Independece of error
Homoscedasticity: contant variance
Error must be normally distributed with mean 0 and variance 1.





(3) What does the coefficient m represent in the equation Y=mX+c?

Ans: m represents slope.





(4) What does the intercept c represent in the equation Y=mX+c?

Ans: c is y-intercept, the value of y when x is zero.





(5) How do we calculate the slope m in Simple Linear Regression?

Ans: slope=cov(x,y)/v(x)





(6) What is the purpose of the least squares method in Simple Linear Regression?

Ans: Lest square is best fit of model, calculated by minimizing the sum of squared difference between observed and predicted values.





(7) How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

Ans: Represents teh proportion of variance in the dependent variable obtained by the independent variable. Higher R^2 indicates better fit.





(8) What is Multiple Linear Regression?

Ans: A statistical model with one dependent variable and more than one independent variable, called as multiple linear regression model.





(9) What is the main difference between Simple and Multiple Linear Regression?

Ans: Both are having only one dependent variable but SLR is having only one independent variable and MLR is having more than one independent varialbe.





(10) What are the key assumptions of Multiple Linear Regression?

Ans: There must be linear relationship between two variables.
Errors must be independent of each other.
variance must be constant.
Eroros must be normaly distributed with mean 0 and variance 1.
There must be high correlation between independent variables.




(11) What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

Ans: Heteroscedasticity is unequal variance and it can lead to unreliable estimates of standard errors and p-values.





(12) How can you improve a Multiple Linear Regression model with high multicollinearity?

Ans: Remove or combine highly correlated independent variables, use dimensionality reduction techniques





(13) What are some common techniques for transforming categorical variables for use in regression models?

Ans: one-hot encoding and dummy coding.





(14) What is the role of interaction terms in Multiple Linear Regression?

Ans: They capture the combined effect of two or more independent variables on the dependent variable.





(15) How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

Ans: In SLR, it's value of Y, when x is 0. In MLR, it's value of y when all x's are 0.





(16) What is the significance of the slope in regression analysis, and how does it affect predictions?

Ans: Slope quantifies the change in y for a unit change in x. It directly affects predictions as it determines teh steepness of the regression line.





(17) How does the intercept in a regression model provide context for the relationship between variables?

Ams: It estiblishes a baseline value for y when independent variables are abset, providing a starting point for understanding the relationship.





(18) What are the limitations of using R² as a sole measure of model performance?

Ans: It doesn't account for model complexity, can be artificially inflated by adding irrelevant variables, and doesn't assess preictive accuracy.





(19) How would you interpret a large standard error for a regression coefficient?

Ans: Indicates greater uncertainty in the estimate of the coefficient, suggesting the variable may not be a srong predictor.





(20) How can heteroscedasticity be identified in residual plots, and why is it important to address it?

Ans: Look for funnel shape in the residual plot. It's important to address becaouse it can lead to inaccurate statistical inference.





(21) What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

Ans: The model may include irrelavant variables that inflate R^2 but doesn't improve the model's predictive power.





(22) Why is it important to scale variables in Multiple Linear Regression?

Ans: To prevent variables with larger scales from dominating the model and to ensure that coefficients are comparable.





(23) What is polynomial regression?

Ans: A form of regression analysis where the relatioship between the independent and dependent variables is modeled as an nth-degree polynomial.





(24) How does polynomial regression differ from linear regression?

Ans: Polynomial regression allows for curved relationships, while linear regression assumes a straight line.





(25) When is polynomial regression used?

Ans: When the data shows a non-linear  relationship between variables.

(28) What are the limitations of polynomial regression?
